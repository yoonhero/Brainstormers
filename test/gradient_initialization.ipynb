```python
import numpy as np
import matplotlib.pyplot as plt 

def sigmoid(x):
    return 1/(1+np.exp(-x))

x = np.random.randn(100)
print(x.shape)
node_num = 100
hidden_layer_size = 5
activations = {}
```

```python
for i in range(hidden_layer_size):
    if i != 0:
        x = activations[i-1]
    w = np.random.randn(node_num)
    a = x*w
    z = sigmoid(a)

    print(z)

    activations[i] = z
```

```python
for i, a in activations.items():
    plt.subplot(1, len(activations), i+1)
    plt.title(str(i+1) + "-layer")
    print(a.flatten())
    plt.hist(a.flatten(), 100, range(-1, 1))

plt.show()
```
